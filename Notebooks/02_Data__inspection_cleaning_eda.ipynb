## Outliers detection

```python
def iqr_outlier_bounds(series, k=1.5):
    q1, q3 = series.quantile(0.25), series.quantile(0.75)
    iqr = q3 - q1
    return q1 - k * iqr, q3 + k * iqr

lower, upper = iqr_outlier_bounds(df["years_experience"])
outliers = df[(df["years_experience"] < lower) | (df["years_experience"] > upper)]
print(f"IQR bounds: [{lower:.2f}, {upper:.2f}]")
print(f"Flagged outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")

plt.figure(figsize=(8, 4))
sns.boxplot(x=df["years_experience"])
plt.title("years_experience — boxplot (IQR outlier view)")
plt.tight_layout()
plt.show()
```

## Data Cleaning

```python
#Imputation of missing values using median
```

```python
cols_to_impute = ["linkedin_profile_score", "competition_awards_count", "certification_prestige_score"]

for col in cols_to_impute:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
    print(f"{col}: imputed {missing[col]} missing values with median = {median_val}")

print("\nRemaining missing values:", df[cols_to_impute].isna().sum().sum())
```

```python
#outliers handling in years_experience column
```

```python
before = df["years_experience"].copy()
df["years_experience"] = df["years_experience"].clip(upper=upper)

print(f"Values capped: {(before != df['years_experience']).sum()}")
print(df["years_experience"].describe())

plt.figure(figsize=(8, 4))
sns.boxplot(x=df["years_experience"])
plt.title("years_experience — after capping at IQR upper bound")
plt.tight_layout()
plt.show()
```

### Multivariet Analysis (Box plot)

#### Observations from Multivariate Analysis (Box Plots):

*   **Selected vs. Rejected Candidates**: Across all analyzed numerical features (`technical_skill_score`, `aptitude_score`, `communication_score`, `interview_score`, `ats_score`, `years_experience`), candidates who were `Selected (1)` generally exhibit higher median scores and tighter distributions compared to `Rejected (0)` candidates. This indicates that higher performance in these areas is associated with a greater likelihood of selection.
*   **Score Distribution for Selected Candidates**: Selected candidates tend to have their scores clustered in the upper quartile, suggesting a consistent high performance profile. The interquartile range (IQR) for selected candidates is often narrower, indicating less variability in their scores.
*   **Score Distribution for Rejected Candidates**: Rejected candidates show a wider spread of scores, with medians often lower than those of selected candidates. They also tend to have more outliers on the lower end, highlighting a broader range of performance levels within this group.
*   **`years_experience`**: Selected candidates generally have more years of experience, with a higher median and potentially a higher upper quartile. However, the presence of outliers in both groups suggests that while experience is a factor, it's not the sole determinant of selection.

#### Conclusion on Multicollinearity

```python
pairs = [
    ("projects_count", "project_quality_score"),
    ("certifications_count", "certification_prestige_score"),
    ("technical_skill_score", "github_coding_profile_score"),
]

for a, b in pairs:
    r = df[a].corr(df[b])
    print(f"{a} vs {b}: r = {r:.3f}")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

vif_df = add_constant(df[numerical_cols])
vif_scores = pd.DataFrame({
    "feature": vif_df.columns,
    "VIF": [variance_inflation_factor(vif_df.values, i) for i in range(vif_df.shape[1])]
})
vif_scores = vif_scores[vif_scores["feature"] != "const"].sort_values("VIF", ascending=False)
vif_scores
```

####  Observations:

Based on the VIF scores, we can confidently state that multicollinearity is not a significant issue among the numerical features in this dataset. This is a positive outcome for subsequent modeling, as it ensures that feature coefficients will be more stable and interpretable.

VIF scores across all 13 numeric features range from ~1.0 to ~1.7 — **all comfortably under
the 5-10 concern threshold**, with no sign of problematic multicollinearity anywhere in the
feature set.
